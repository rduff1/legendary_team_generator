```
# Determine powerlvl of combination

def determine_powerlevel(combination):
  total = 0
  
  #Mastermind
  curr_lvl = 1
  multiplier = 40 / 8
  for a in pl_masterminds:
    if combination.mastermind.name in a:
      total += curr_lvl * multiplier
    curr_lvl += 1
    
  #Scheme
  curr_lvl = 1
  multiplier = 30 / 5
  found = False
  for a in pl_schemes:
    if combination.scheme.name in a:
      total += curr_lvl * multiplier
      found = True
    curr_lvl += 1
  if found == False:
    total += 3 * multiplier
    
  #Villains
  curr_lvl = 1
  multiplier = 20 / 6 / len(combination.villains)
  for a in pl_villains:
    for b in combination.villains:
      if b in a:
        total += curr_lvl * multiplier
    curr_lvl += 1
    
  #Henchmen
  curr_lvl = 1
  multiplier = 10 / 4 / len(combination.henchmen)
  for a in pl_henchmen:
    for b in combination.henchmen:
      if b in a:
        total += curr_lvl * multiplier
    curr_lvl += 1
    
  return total
      
      
```
